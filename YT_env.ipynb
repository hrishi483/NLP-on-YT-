# Function to get channel Captions

```python
def get_channel_stats(youtube,channel_ids):
    requests=youtube.channels().list(
                part='snippet,contentDetails,statistics',
                id=channel_ids)

    response=requests.execute()
    channel_info=[]
    for i in range(len(response['items'])):
        data=dict(Channel_name=response['items'][i]['snippet']['title'],
                  Subscribers=response['items'][i]['statistics']['subscriberCount'],
                 Views=response['items'][i]['statistics']['viewCount'],
                 Total_videos=response['items'][i]['statistics']['videoCount'],
                 Desc=response['items'][i]['snippet']['description'],
                 Published_Date=response['items'][i]['snippet']['publishedAt'],
                 thumbnail=response['items'][i]['snippet']['thumbnails']['high']['url'])
        channel_info.append(data)



    return channel_info
```

```python
def get_captions(video_ids):
    for video_id in video_ids:
        response=YouTubeTranscriptApi.get_transcript(
            video_id, languages=['en'])

        text_to_summarize=[]
        for r in response:
            text_to_summarize.append(r['text'])

#         text_to_summarize = "Your input text here"
#         summary = generate_summary(text_to_summarize)
        return text_to_summarize
    print('\n','*'*100,'\n')

```

```python
def summarize_text(text):
    # print(type(text))
    sentences=text

    new_text=[]
    #Removing the stopwords from the sentences.
    for sentence in sentences:
        words=sentence.split(' ')
        new_sentence=""
        for word in words:
            if word not in stopwords.words("english"):
                word = lemmatizer.lemmatize(word)
                new_sentence+=" "+word
        new_sentence=new_sentence.replace(',','')
        new_text.append(new_sentence)


    #Calculate frequency of each word in the sentence
    word_freq=dict()
    for sentence in new_text:
        words=sentence.split(' ')
        for word in words:
            if (word not in word_freq.keys()) & (len(word)>2):
                word_freq[word]=1
            elif (len(word)>2):
                word_freq[word]+=1

    max_freq=max(word_freq.values())


    sentence_val_dict=dict()
    for sentence in sentences:
        words=sentence.split(' ')
        sentence_val=0

        for word in words:
            sentence_val+=float((word_freq['word'])/(max_freq))
        sentence_val_dict[sentence]=sentence_val

    summarized_text=[]
    best_score=np.mean(list(sentence_val_dict.values()))


    for sentence,score in sentence_val_dict.items():
        if score>best_score:
            summarized_text.append(sentence)
    # print(sentence_val_dict)
    print("Length of summarized Text = ",len(summarized_text),"Length of Text = ",len(text))
    print("Summarized text is ",(100*len(summarized_text))/len(text),"% of orignal text")
    '.'.join(summarized_text)
    return summarized_text
```

```python


#Directly using NLTK Library:
def summarize_text(text, num_sentences):
    nlp = spacy.load('en_core_web_sm')

    doc = nlp(" ".join(text))

    ranked_sentences = sorted(doc.sents, key=lambda s: len(s), reverse=True)

    selected_sentences = ranked_sentences[:num_sentences]

    summary = ' '.join([str(sentence) for sentence in selected_sentences])

    return summary

# Example usage
text = speech
num_sentences = int(0.001*len(text))
print(len(text),num_sentences)
# pprint(text)
summary = summarize_text(text, num_sentences)
pprint(summary)
```
